Data Fields
from [Bike Sharing Demand | Kaggle](https://www.kaggle.com/c/bike-sharing-demand/data)


### Description
* datetime - hourly date + timestamp  
* season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
* holiday - whether the day is considered a holiday
* workingday - whether the day is neither a weekend nor holiday
* weather 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy 
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist 
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds 
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
* temp - temperature in Celsius
* atemp - "feels like" temperature in Celsius
* humidity - relative humidity
* windspeed - wind speed
* casual - number of non-registered user rentals initiated
* registered - number of registered user rentals initiated
* count - number of total rentals

### Evaluation
Submissions are evaluated one the Root Mean Squared Logarithmic Error (RMSLE).

## 추가전처리

```python
train.head()
```

```python
train = pd.get_dummies(train, columns = ['season', 'weather', 'dayofweek'], drop_first= True)
```

```python
# def concatenate_year_month(datetime):
#     return "{0}-{1}".format(datetime.year, datetime.month)

# train["year_month"] = train["datetime"].apply(concatenate_year_month)

def yearMonthChange(yearMonth):
    if len(yearMonth) == 6:
        new_yearMonth = yearMonth[:4] +'0'+ yearMonth[5]
        return int(new_yearMonth)
    else:
        new_yearMonth = yearMonth[:4] + yearMonth[5:7]
        return int(new_yearMonth)

train['year_month'] = train['year_month'].apply(yearMonthChange)
```

```python
train.drop(['minute', 'second'], axis=1, inplace = True)
train.columns
```

```python
features = ['holiday', 'workingday', 'day','atemp', 'humidity','windspeed', 'casual',
       'hour', 'year_month', 'season_2', 'season_3', 'season_4',
       'weather_2', 'weather_3', 'weather_4', 'dayofweek_1', 'dayofweek_2', 'dayofweek_3',
       'dayofweek_4', 'dayofweek_5', 'dayofweek_6']
scale_features = ['atemp', 'humidity','windspeed', 'casual']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train[scale_features] = scaler.fit_transform(train[scale_features])
train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
          train[features], train['count'], test_size=0.33, random_state=42)
```

```python
LR = LinearRegression()
LR.fit(train[features], train['count'])
LR.score(X_train, y_train)
LR.score(X_test, y_test)
# LR.score(train[features], train['count'])
```
