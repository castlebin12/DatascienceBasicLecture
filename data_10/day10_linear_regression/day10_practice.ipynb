# Statistical Data Analysis Using ML

### 데이터 로드

```python
import pandas as pd
df = pd.read_csv('../day10_data/Advertising.csv')
```

```python
# 데이터 확인
df.head()
```

### Train/Test Set 분할

보통은 실제 데이터를 바탕으로 Train을 하고, 새로운 데이터를 기반으로 Test를 하지만, 본 연습에서는 주어진 데이터를 Train 데이터와 Test 데이터로 나누어 분석하고자 한다. 이를 위해 Scikit-Learn이 제공하는 train_test_split 모듈을 사용한다.

```python
from sklearn.model_selection import train_test_split

#아래에 코드 작성
features = ['TV', 'radio', 'newspaper']
X_train, X_test , y_train, y_test = train_test_split(df[features], df['sales'], random_state =2 )

print(X_train)
print(y_train)
```

### Linear regression

#### Linear regression 수행

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from matplotlib import pyplot as plt
```

```python
#아래에 코드 작성
lm = linear_model.LinearRegression()
lm.fit(X_train, y_train)
```

#### 통계값 계산

```python
# The coefficients
print('Coefficients : {}' .format(lm.coef_))
# The mean squared error
print('Mean squared error : {}' .format(mean_squared_error(y_train, lm.predict(X_train))))
# Explained variance score: 1 is perfect prediction
print('R^2: {}'.format(r2_score(y_train, lm.predict(X_train))))
# Accuracy
print('Training set score : {}' .format(lm.score(X_train, y_train)))
print('Test set score : {}' .format(lm.score(X_test, y_test)))
```

#### 전체 판매량 재예측 및 시각화

학습된 lm 모델을 이용하여 전체 데이터를 다시 predict

```python
#아래에 코드 작성
X = df[features]
X.head()
```

```python
pred = lm.predict(X)
df['Predict'] = pred
df.head()
```

sales와 predict 그래프 그리기

```python
#아래에 코드 작성
df.plot(y=['sales', 'Predict'], figsize=(30,10), linewidth=5)
print(df.shape)
```
