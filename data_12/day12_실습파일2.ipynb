## 모델 돌려보기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Immunotherapy.csv')
data['sex'] = data['sex'].replace('male',0)
data['sex'] = data['sex'].replace('female',1)
data.head()



sum = 0
cnt = 0
for i,j in enumerate(data['age']):
    if j.isdigit() == True:
        sum += int(j)
        cnt += 1

age_avg = int(sum/cnt)

# for i,j in enumerate(data['age']):
#     if j.isdigit() == False:
#         print(data.loc[i, 'age'])
#         data.loc[i, 'age'] = age_avg
#         #print(i, data.loc['age',i])

# data['age'].isnumeric()
# data[type(data['age'])]
# data.loc[type('age') != int] = '?'
data.loc[data['age'].apply(lambda x: not x.isdigit()), 'age'] = age_avg

data.head(11)

data.isna()

data['Number_of_Warts'].isna()

num_of_warts_avg = data.loc[data['Number_of_Warts'].notnull(), 'Number_of_Warts'].mean()
data['Number_of_Warts'].fillna(num_of_warts_avg , inplace = True)
```

```python
# feature = data.f
features = list(data.columns)
features.pop()
x_train, x_test, y_train, y_test = train_test_split(data[features],data['Result_of_Treatment'], random_state =2 )
```

## 1. KNN

```python
knn = KNeighborsClassifier(n_neighbors = 3).fit(x_train,y_train)
```

```python
train_score = []
test_score = []
neighbors_settings = range(1,11)

for i in neighbors_settings:
    clrKN = KNeighborsClassifier(n_neighbors = i)
    train_knn = clrKN.fit(x_train,y_train)
    train_score.append(train_knn.score(x_train, y_train))
    test_score.append(train_knn.score(x_test, y_test))

plt.figure(figsize=(10,10))
plt.plot(neighbors_settings, train_score , 'b', label = 'train')
plt.plot(neighbors_settings, test_score, 'r', label = 'test')
plt.legend()
plt.show()
```
