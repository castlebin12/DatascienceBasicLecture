# load IRIS dataset

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt2
# from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
iris = load_iris()

y = iris.target

data = pd.DataFrame(iris.data)
data.columns=['Sepal_length','Sepal_width','Petal_length','Petal_width']

data.head()
```

## Sepal_length, Sepal_width 두 가지 feature를 사용해서
# k-Means clustering

```python
from sklearn.cluster import KMeans
```

```python
feature = ['Sepal_length', 'Sepal_width']
X = data[feature]
X
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

```python
# import library for 3D plotting
from mpl_toolkits import mplot3d

# magic function for interactive plot
%matplotlib notebook

plt.figure('3D Test', figsize=(7,5))
ax = plt.axes(projection = '3d')

ax.scatter(X['Sepal_length'], X['Sepal_width'],X['Petal_length'], c=y_pred)
# ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],kmeans.cluster_centers_[:, 2])
```

```python
plt.scatter(X['Sepal_length'], X['Sepal_width'], c=y_pred)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='^', c=[1,2,0], s=150, linewidth=2, edgecolors='black')
plt.xlabel("Sepal_length")
plt.ylabel("Sepal_width")
print(kmeans.cluster_centers_[:, 0])
print(kmeans.cluster_centers_[:, 1])
# print(kmeans.cluster_centers_[:, 2])
kmeans.cluster_centers_[:, 0].shape
```
